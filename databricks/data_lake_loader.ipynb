# Users

```
data_path = f"abfss://bronze@{demo_storage_account_name}.dfs.core.windows.net/users/"
checkpoint_path = f"abfss://bronze@{demo_storage_account_name}.dfs.core.windows.net/_checkpoint/users"

(spark.readStream
  .format("cloudFiles")
  .option("cloudFiles.format", "json")
  .option("cloudFiles.schemaLocation", checkpoint_path)
  .load(data_path)
  .writeStream
  .option("checkpointLocation", checkpoint_path)
  .trigger(availableNow=True)
  .toTable("mycatalog.mydb.users"))
```
